```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.DataFrame(iris.data)
data
```

```python
data["target"] = iris.target
```

```python
data
```

```python
X = data.drop("target", axis = "columns")
y = data.target
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier (n_estimators = 15)
model.fit(X_train, y_train)
```

```python
y_test
```

```python
X_test
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
y_predicted
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
dir(diabetes)
```

```python
df = pd.DataFrame(diabetes.data)
df
```

```python
df["target"] = diabetes.target
df
```

```python
independent = df.drop("target", axis = "columns")
independent
```

```python
dependent = df.target
dependent
```

```python
from sklearn.model_selection  import train_test_split
independent_train, independent_test, dependent_train, dependent_test = train_test_split(independent, dependent, test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier (n_estimators = 15)
model.fit(independent_train, dependent_train)
```

```python
independent_test
```

```python
dependent_test
```

```python
model.score(independent_train, dependent_train)
```

```python
dependent_predicted = model.predict(independent_test)
dependent_predicted
```
